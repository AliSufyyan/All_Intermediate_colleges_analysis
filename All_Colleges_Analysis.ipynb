```python
Rating_count=df['Rating'].value_counts()
```

```python
Rating_count=pd.DataFrame(Rating_count)
```

```python
Rating_count.reset_index()
```

```python
df['Rating']=df['Rating'].str.replace('*','', regex=False)
```

```python
df['Rating']=pd.to_numeric(df['Rating'],errors='coerce')
```

```python
low_ratings=df[(df['Rating']<2.5) & (df['Rating']>0)]
```

```python
low_ratings
```

```python
#plt.figure(figsize=(6,4), dpi=200)
sns.countplot(x='Sector', data=low_ratings, hue='Sector')
```
